### Data Cleaning

```python
df = merged_df
```

Handling Null Values

```python
# Check for null values
df.isnull().sum()[df.isnull().sum() > 0]
```

We have to handle these null values as such -
1. device_access: The best approach is to create a new category 'unknown' instead of guessing the mode.
2. extracurricular: It is null possibly because students who do not have opted for it simply left it blank. Thus, we can fill it with 'None'
3. admission_year_predicted: This has a lot of null values and this feature is not reliable enough for analysis or training the model. So, we will drop the entire column.
4. scholarships_available: It is null possibly because the college doesn't has scholarships. So, we can fill it with 'None', assuming it doesn't have any scholarship.

```python
df['device_access'].fillna('Unknown', inplace=True)
df['extracurricular'].fillna('None', inplace=True)
df['scholarships_available'].fillna('None', inplace=True)
df.drop(columns=['admission_year_predicted'], axis=1, inplace=True, errors='ignore')

# List of critical numerical columns for all models that might have NaNs
critical_numerical_cols = [
    'quantitative_score', 'logical_score', 'verbal_score', 'creative_score',
    'technical_score', 'aggregate_percentage', 'last_year_cutoff',
    'dropout_risk_score'
]

# Convert these columns to numeric, coercing errors to NaN
# This ensures that any non-numeric missing values (like 'None' or empty strings) become actual np.nan
for col in critical_numerical_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Drop rows where any of these critical numerical columns have NaNs
# This should effectively remove the problematic row(s) early in the cleaning process
initial_rows = df.shape[0]
df.dropna(subset=critical_numerical_cols, inplace=True)
rows_after_drop = df.shape[0]
print(f"Dropped {initial_rows - rows_after_drop} rows due to NaNs in critical numerical columns.")
```

##### 5. Data Processing for ML
We will prepare the data specifically for our first model: the Dropout Risk Prediction Model. The goal is to predict which students are at a high risk of dropping out.

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
# Define the features (X) we will use for prediction
feature_columns = [
    'aggregate_percentage',
    'score_cutoff_difference', # Our powerful engineered feature
    'socio_economic_status',
    'parental_education',
    'school_type',
    'internet_access',
    'device_access',
    'quantitative_score',
    'logical_score',
    'verbal_score'
]
```

```python
# Define the target (y) we want to predict
target_column = 'dropout_risk_score'
```

```python
# Create a new, smaller DataFrame for this specific model
model_df = merged_df[feature_columns + [target_column]].copy()
```

```python
# Recreate model_df to ensure target_column is present in case of re-execution
model_df = merged_df[feature_columns + [target_column]].copy()

print("--- 2. Creating Binary Target Variable 'high_dropout_risk' ---")
risk_threshold = 70
model_df['high_dropout_risk'] = (model_df[target_column] > risk_threshold).astype(int)

# Drop the original score column as we now have our target
model_df.drop(target_column, axis=1, inplace=True)

# Remove any rows with NaN values after creating the target variable
model_df.dropna(inplace=True)

# Display the class balance
print(f"Risk Threshold set at: {risk_threshold}")
print("Distribution of target variable:")
print(model_df['high_dropout_risk'].value_counts(normalize=True))
```

```python
# We convert all text-based columns into a numerical format.

print("\n--- 3. Encoding Categorical Features ---")

# Identify which columns are categorical
categorical_features = model_df.select_dtypes(include=['object']).columns

# Use pandas.get_dummies for one-hot encoding
model_df_encoded = pd.get_dummies(model_df, columns=categorical_features, drop_first=True)

print("Data has been successfully encoded.")
print(f"Original number of features: {len(feature_columns)}")
print(f"Number of features after encoding: {len(model_df_encoded.columns) - 1}") # -1 for the target column
```

```python
# Separate the features (X) from the target (y)

X = model_df_encoded.drop('high_dropout_risk', axis=1)
y = model_df_encoded['high_dropout_risk']

print("\n--- 4. Final X (features) and y (target) are ready for model training ---")
print("Shape of X:", X.shape)
print("Shape of y:", y.shape)
```

```python
# We'll use an 80/20 split. 80% for training, 20% for testing.
# `random_state` ensures that we get the same split every time we run the code.
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y  # 'stratify=y' ensures the proportion of high/low risk students is the same in both train and test sets.
)

print("--- Data Splitting Complete ---")
print("Shape of X_train:", X_train.shape)
print("Shape of y_train:", y_train.shape)
print("---")
print("Shape of X_test:", X_test.shape)
print("Shape of y_test:", y_test.shape)
```

##### 6. Model training


```python
# Import the models and evaluation metrics
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from lightgbm import LGBMClassifier
from sklearn.metrics import classification_report, accuracy_score
from sklearn.preprocessing import StandardScaler
```

```python
# It's a best practice to scale numerical features for these models.

# --- NEW: Explicit NaN check and removal before scaling ---
print("\n--- Performing final NaN check before scaling ---")
if X_train.isnull().sum().sum() > 0:
    print(f"Warning: Found {X_train.isnull().sum().sum()} NaNs in X_train. Dropping rows with NaNs.")
    # Drop NaNs from X_train and corresponding y_train
    nan_rows_train = X_train.isnull().any(axis=1)
    X_train = X_train[~nan_rows_train]
    y_train = y_train[~nan_rows_train]
    print(f"X_train shape after dropping NaNs: {X_train.shape}")
    print(f"y_train shape after dropping NaNs: {y_train.shape}")

if X_test.isnull().sum().sum() > 0:
    print(f"Warning: Found {X_test.isnull().sum().sum()} NaNs in X_test. Dropping rows with NaNs.")
    # Drop NaNs from X_test and corresponding y_test
    nan_rows_test = X_test.isnull().any(axis=1)
    X_test = X_test[~nan_rows_test]
    y_test = y_test[~nan_rows_test]
    print(f"X_test shape after dropping NaNs: {X_test.shape}")
    print(f"y_test shape after dropping NaNs: {y_test.shape}")
else:
    print("No NaNs found in X_train or X_test.")


scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

**Models**

In this project, we have two main models:

1.  **The Dropout Risk Prediction Model (which we just built).**
    *   **Input:** Student academic and demographic data.
    *   **Output:** A prediction (Yes/No) if the student is at high risk of dropping out.
    *   **Purpose:** Powers proactive student tracking system (Outcome 4).

2.  **The Stream Recommendation Model (which we will build next).**
    *   **Input:** Student's aptitude quiz scores.
    *   **Output:** A prediction of the most suitable stream (Science, Commerce, Arts, etc.).
    *   **Purpose:** Powers personalized career guidance system (Outcome 1).

###### a. Dropout Risk Prediction Model

```python
# Define the models we want to evaluate
models = {
    "Logistic Regression": LogisticRegression(random_state=42),
    "Random Forest": RandomForestClassifier(random_state=42),
    "LightGBM": LGBMClassifier(random_state=42)
}
```

```python
# Loop through each model to train, predict, and evaluate
for model_name, model in models.items():

    # Train the model
    print(f"--- Training {model_name} ---")
    model.fit(X_train_scaled, y_train)
    print("Training complete.")

    # Make predictions on the test set
    y_pred = model.predict(X_test_scaled)

    # Evaluate the model
    print(f"\n--- Evaluation for {model_name} ---")

    # Calculate accuracy
    accuracy = accuracy_score(y_test, y_pred)
    print(f"Accuracy: {accuracy:.4f}")

    # Print the detailed classification report
    # This shows precision, recall, and f1-score for each class (0 and 1)
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred, target_names=["Not High Risk (0)", "High Risk (1)"]))
    print("----------------------------------------------------\n")
```

**Model Analysis and Selection**
1. **Logistic Regression**:
- This was a decent baseline, but its Recall (0.52) for high-risk students is too low. This means it failed to identify almost half of the students who are actually at risk, which is a critical failure for our project's goal.

2. **Random Forest & LightGBM**:
- These two models performed identically and exceptionally well.
- They both achieved a high F1-Score (0.79) for the "High Risk" class, which is a strong indicator of a balanced and effective model.
- Their Recall (0.77) means they successfully identified 77% of the students who were truly at risk, a huge improvement over Logistic Regression.
- Their Precision (0.82) means that when they flag a student, they are correct 82% of the time.

Thus, We will select **Random Forest** model for Dropout Risk Prediction!
